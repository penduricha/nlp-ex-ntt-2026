```python
# Câu 1
def generate_logical_form_13_1():
    # Giả sử parser đã trích xuất:
    raw_rels = {
        "PRED": "BUY",
        "LOBJ": "TICKET1",
        "AT": "THEATER1"
    }
    
    # Áp dụng các mẫu từ Hình 13.2 [cite: 10]
    logical_form = {
        "PRED": raw_rels["PRED"],
        "THEME": raw_rels["LOBJ"],
        "LOC": raw_rels["AT"]
    }
    return logical_form

print(generate_logical_form_13_1())
```

```python
# Câu 2
def semantic_interpreter(verb, lobj, beneficiary):
    return {
        "ACTION": verb,
        "THEME": lobj,
        "BENEFICIARY": beneficiary
    }
print(semantic_interpreter("BUY", "TICKET", "ME"))
```

```python
# Câu 3
def represent_quantifiers(noun, quantifier, action):
    # Collective: Seven men together
    collective = f"({action} (GROUP {quantifier} {noun}))"
    # Distributive: Each of the seven men met (ít hợp lý với từ 'met')
    distributive = f"(FORALL x IN {quantifier} {noun}: {action}(x))"
    return {"Collective": collective, "Distributive": distributive}

print(represent_quantifiers("MEN", "7", "MET"))
```
