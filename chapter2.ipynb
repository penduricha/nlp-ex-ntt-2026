```python
import pandas as pd
```

```python
def min_edit_distance(source, target, ins_cost=1, del_cost=1, sub_cost=1):
    n = len(source)
    m = len(target)
    
    # Tạo ma trận D[n+1][m+1] [cite: 158, 165]
    # D[i, j] là khoảng cách giữa i ký tự đầu của source và j ký tự đầu của target
    D = [[0] * (m + 1) for _ in range(n + 1)]
    
    # Trường hợp cơ sở: chuỗi rỗng [cite: 160, 161]
    for i in range(1, n + 1):
        D[i][0] = D[i-1][0] + del_cost
    for j in range(1, m + 1):
        D[0][j] = D[0][j-1] + ins_cost
        
    # Tính toán ma trận từ dưới lên [cite: 159, 162]
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if source[i-1] == target[j-1]:
                cost = 0 # Ký tự giống nhau, chi phí thay thế = 0 [cite: 154]
            else:
                cost = sub_cost
            
            D[i][j] = min(
                D[i-1][j] + del_cost,    # Xóa (Deletion) 
                D[i][j-1] + ins_cost,    # Chèn (Insertion) 
                D[i-1][j-1] + cost       # Thay thế (Substitution) 
            )
            
    # Hiển thị ma trận dưới dạng bảng cho đẹp (giống lưới trong slide)
    df = pd.DataFrame(
        D, 
        index=['#'] + list(source), 
        columns=['#'] + list(target)
    )
    return D[n][m], df
```

```python
dist_22, matrix_22 = min_edit_distance("leda", "deal")
print("Câu 2.2: Khoảng cách từ 'leda' đến 'deal' là:", dist_22)
print(matrix_22)
```

```python
dist_brief, _ = min_edit_distance("drive", "brief")
dist_divers, _ = min_edit_distance("drive", "divers")

print(f"Câu 2.3:")
print(f"Khoảng cách 'drive' -> 'brief': {dist_brief}")
print(f"Khoảng cách 'drive' -> 'divers': {dist_divers}")

if dist_brief < dist_divers:
    print("=> 'drive' gần với 'brief' hơn.")
elif dist_brief > dist_divers:
    print("=> 'drive' gần với 'divers' hơn.")
else:
    print("=> 'drive' có khoảng cách bằng nhau tới cả hai từ.")
```
