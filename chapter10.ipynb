```python
# Định nghĩa văn phạm
grammar = {
    'S': [['NP', 'VP']],
    'NP': [['ART', 'N'], ['ART', 'ADJ', 'N']],
    'VP': [['V'], ['V', 'NP']]
}
```

```python
def top_down_parse(symbols, tokens, depth=0):
    indent = "  " * depth
    if not symbols and not tokens:
        return True
    if not symbols or not tokens:
        return False
    
    first = symbols[0]
    
    # Nếu là ký hiệu kết thúc (từ loại)
    if first in ['ART', 'N', 'V', 'ADJ']:
        if first == tokens[0]:
            print(f"{indent}Khớp: {first} với {tokens[0]}")
            return top_down_parse(symbols[1:], tokens[1:], depth + 1)
        return False
    
    # Nếu là ký hiệu không kết thúc (vế trái của luật)
    for rule in grammar.get(first, []):
        print(f"{indent}Thử luật: {first} -> {' '.join(rule)}")
        if top_down_parse(rule + symbols[1:], tokens, depth + 1):
            return True
    return False
```

```python
# Ví dụ cho câu (i): ART N V ART ADJ N
print("Phân tích câu (i):")
top_down_parse(['S'], ['ART', 'N', 'V', 'ART', 'ADJ', 'N'])
```

```python
class MaltParserSim:
    def __init__(self, sentence):
        self.stack = ["ROOT"]
        self.buffer = sentence
        self.arcs = []

    def shift(self):
        word = self.buffer.pop(0)
        self.stack.append(word)
        print(f"SHIFT: Stack={self.stack}, Buffer={self.buffer}")

    def right_arc(self, relation):
        head = self.stack[-1]
        dep = self.buffer[0]
        self.arcs.append(f"{relation}({head}, {dep})")
        self.stack.append(self.buffer.pop(0))
        print(f"RIGHT-ARC: {relation}({head}, {dep})")

    # Giả lập quá trình phân tích một phần câu (a)
    def parse(self):
        print("Bắt đầu phân tích...")
        self.shift() # ROOT
        self.shift() # He
        self.right_arc("nsubj") # books(He)
        # ... tiếp tục các bước tương tự cho đến khi buffer rỗng
        print(f"Các quan hệ đã tạo: {self.arcs}")

parser = MaltParserSim(["He", "books", "me", "the", "morning", "flight"])
parser.parse()
```
