Cấu trúc từ điển và văn phạm

```python
# Cấu trúc từ điển mở rộng
lexicon = {
    "He": {"CAT": "PRO", "AGR": "3s"},
    "was": {"CAT": "V", "ROOT": "BE", "VFORM": "past", "AGR": "3s", "SUBCAT": "adj"},
    "sad": {"CAT": "ADJ"},
    "to": {"CAT": "TO"},
    "see": {"CAT": "V", "ROOT": "SEE", "VFORM": "base", "SUBCAT": "np"},
    "the": {"CAT": "ART", "AGR": "3s"},
    "dog": {"CAT": "N", "ROOT": "DOG", "AGR": "3s"},
    "cry": {"CAT": "V", "ROOT": "CRY", "VFORM": "base", "SUBCAT": "none"}
}

# Văn phạm mở rộng cho các cấu trúc bổ ngữ
grammar_rules = [
    ("S", ["NP", "VP"]),
    ("VP", ["V", "ADJP"]),      # Ví dụ: "was sad"
    ("ADJP", ["ADJ", "VP"]),    # Ví dụ: "sad to see..."
    ("VP", ["TO", "VP"]),       # Ví dụ: "to see..."
    ("VP", ["V", "NP"])         # Ví dụ: "see the dog"
]
```

```python
def check_agreement(subject, verb, object_):
    if subject['AGR'] != verb['AGR']:
        return "Lỗi đồng thuận Subject-Verb"
    if subject['CASE'] != 'nom':
        return "Lỗi: Chủ ngữ không ở dạng Nominative"
    if object_['CASE'] != 'acc':
        return "Lỗi: Tân ngữ không ở dạng Accusative"
    return "Câu hợp lệ"

# Kiểm tra "I hit him"
subject_i = {"CAT": "PRO", "AGR": "1s", "CASE": "nom"}
verb_hit = {"CAT": "V", "AGR": "1s"}
object_him = {"CAT": "PRO", "AGR": "3s", "CASE": "acc"}

print(check_agreement(subject_i, verb_hit, object_him)) 
```
