```python
import math
w = [0.3, 0.5, 0.8]
b = 0.5
x = [0.6, 0.5, 0.2]
```

```python
# Tính tổng z = w*x + b
z = sum(wi * xi for wi, xi in zip(w, x)) + b

# Hàm kích hoạt sigmoid
def sigmoid(z):
    return 1 / (1 + math.exp(-z))

# Tính kết quả đầu ra y
y = sigmoid(z)
```

```python
print(f"Giá trị z: {z}")
print(f"Kết quả đầu ra y: {round(y, 2)}")
```
